```python
# PrÃ©paration des donnÃ©es
X = df.iloc[:, :-1].values  # Toutes les colonnes sauf la derniÃ¨re (label)
y = df.iloc[:, -1].values   # DerniÃ¨re colonne (label)
```

```python
from sklearn.preprocessing import StandardScaler
# Standardisation des caractÃ©ristiques
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
# SÃ©paration en ensembles d'entraÃ®nement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
print(f"Forme de X_train: {X_train.shape}")
print(f"Forme de X_test: {X_test.shape}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle
# EntraÃ®nement du modÃ¨le Gradient Boosting
gb_classifier = GradientBoostingClassifier(
    n_estimators=100,     # Nombre d'arbres
    learning_rate=0.1,    # Taux d'apprentissage
    max_depth=5,          # Profondeur maximale des arbres
    random_state=42
)
# EntraÃ®ner le modÃ¨le
gb_classifier.fit(X_train, y_train)

```

```python
# PrÃ©dictions sur l'ensemble de test
y_pred = gb_classifier.predict(X_test)

# Ãvaluation du modÃ¨le
accuracy = accuracy_score(y_test, y_pred)
print(f"\nPrÃ©cision du modÃ¨le: {accuracy:.4f}")
```

```python
# Rapport de classification dÃ©taillÃ©
print("\nRapport de classification:")
print(classification_report(y_test, y_pred))
```

```python
# Matrice de confusion
plt.figure(figsize=(14, 12))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.title('Matrice de confusion')
plt.xlabel('PrÃ©dictions')
plt.ylabel('Valeurs rÃ©elles')
plt.tight_layout()
plt.savefig('confusion_matrix.png')
plt.show()
```

```python
# Importance des caractÃ©ristiques
feature_names = df.columns[:-1]
importances = gb_classifier.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
plt.title('Importance des caractÃ©ristiques')
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), [feature_names[i] for i in indices], rotation=90)
plt.tight_layout()
plt.savefig('feature_importance.png')
plt.show()
```

```python
# Visualisation des performances du modÃ¨le
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle
```

```python
# Convertir les labels en format binaire (one-vs-rest)
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]

# Calculer les probabilitÃ©s pour chaque classe
y_score = gb_classifier.predict_proba(X_test)

# Calculer la courbe ROC et l'AUC pour chaque classe
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
# Afficher les courbes ROC pour les 5 premiÃ¨res classes (pour la lisibilitÃ©)
plt.figure(figsize=(10, 8))
colors = cycle(['blue', 'red', 'green', 'orange', 'purple'])
for i, color in zip(range(5), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'ROC curve de {np.unique(y)[i]} (area = {roc_auc[i]:.2f})')
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')
plt.title('Courbes ROC pour diffÃ©rentes cultures (5 premiÃ¨res)')
plt.legend(loc="lower right")
plt.savefig('roc_curves.png')
plt.show()

```

```python
# Fonction de prÃ©diction pour de nouvelles donnÃ©es
def predict_crop(N, P, K, temperature, humidity, ph, rainfall):
    # CrÃ©er un tableau avec les valeurs d'entrÃ©e
    input_data = np.array([[N, P, K, temperature, humidity, ph, rainfall]])
    
    # Standardiser les donnÃ©es d'entrÃ©e
    input_data_scaled = scaler.transform(input_data)
    
    # Faire la prÃ©diction
    prediction = gb_classifier.predict(input_data_scaled)
    
    # Obtenir les probabilitÃ©s pour chaque classe
    probabilities = gb_classifier.predict_proba(input_data_scaled)
    
    # Trouver l'indice de la classe prÃ©dite
    predicted_idx = np.argmax(probabilities)
    
    return prediction[0], probabilities[0][predicted_idx]

# Exemple de prÃ©diction
N = 90
P = 42
K = 43
temperature = 21
humidity = 82
ph = 6.5
rainfall = 200

predicted_crop, confidence = predict_crop(N, P, K, temperature, humidity, ph, rainfall)
print(f"\nPour les valeurs d'entrÃ©e N={N}, P={P}, K={K}, tempÃ©rature={temperature}, humiditÃ©={humidity}, pH={ph}, prÃ©cipitations={rainfall}")
print(f"La culture prÃ©dite est: {predicted_crop} avec une confiance de {confidence:.4f}")
```

```python
import os
# SÃ©rialisation du modÃ¨le et du scaler
model_dir = 'model'
os.makedirs(model_dir, exist_ok=True)

# Enregistrer le modÃ¨le
with open(os.path.join(model_dir, 'gradient_boosting_model.pkl'), 'wb') as f:
    pickle.dump(gb_classifier, f)

# Enregistrer le scaler
with open(os.path.join(model_dir, 'scaler.pkl'), 'wb') as f:
    pickle.dump(scaler, f)

# Enregistrer les noms de colonnes et autres mÃ©tadonnÃ©es
metadata = {
    'feature_names': list(feature_names),
    'target_name': 'label',
    'classes': list(np.unique(y))
}

with open(os.path.join(model_dir, 'metadata.pkl'), 'wb') as f:
    pickle.dump(metadata, f)

print(f"\nModÃ¨le, scaler et mÃ©tadonnÃ©es sauvegardÃ©s dans le dossier '{model_dir}'")
```
