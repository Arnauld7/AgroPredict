Script to get weather and predict based on real time updates

```python
# deployScript.ipynb

import numpy as np
import pandas as pd
import pickle
import os
import joblib
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

class CropPredictor:
    """
    Une classe pour charger et utiliser différents modèles de prédiction de cultures
    """
    
    def __init__(self):
        # Modèles disponibles
        self.models = {
            'gradient_boosting': None,
            'tensorflow': None,
            'tensorflow_lite': None
        }
        
        # Métadonnées associées
        self.metadata = {
            'gradient_boosting': None,
            'tensorflow': None
        }
        
        # Scalers pour la standardisation
        self.scalers = {
            'gradient_boosting': None,
            'tensorflow': None
        }
        
        # Tentative de chargement des modèles disponibles
        self._load_models()
    
    def _load_models(self):
        """Charge tous les modèles disponibles"""
        # Gradient Boosting (scikit-learn)
        try:
            gb_model_path = os.path.join('model', 'gradient_boosting_model.pkl')
            scaler_path = os.path.join('model', 'scaler.pkl')
            metadata_path = os.path.join('model', 'metadata.pkl')
            
            if os.path.exists(gb_model_path) and os.path.exists(scaler_path) and os.path.exists(metadata_path):
                # Charger le modèle
                with open(gb_model_path, 'rb') as f:
                    self.models['gradient_boosting'] = pickle.load(f)
                
                # Charger le scaler
                with open(scaler_path, 'rb') as f:
                    self.scalers['gradient_boosting'] = pickle.load(f)
                
                # Charger les métadonnées
                with open(metadata_path, 'rb') as f:
                    self.metadata['gradient_boosting'] = pickle.load(f)
                
                print("Modèle Gradient Boosting chargé avec succès.")
            else:
                print("Modèle Gradient Boosting ou ses fichiers associés non trouvés.")
        except Exception as e:
            print(f"Erreur lors du chargement du modèle Gradient Boosting: {e}")
        
        # TensorFlow
        try:
            tf_model_path = os.path.join('tf_model', 'best_model.h5')
            tf_metadata_path = os.path.join('tf_model', 'metadata.pkl')
            
            if os.path.exists(tf_model_path) and os.path.exists(tf_metadata_path):
                # Charger le modèle
                self.models['tensorflow'] = tf.keras.models.load_model(tf_model_path)
                
                # Charger les métadonnées (contient aussi le scaler)
                with open(tf_metadata_path, 'rb') as f:
                    self.metadata['tensorflow'] = pickle.load(f)
                
                # Récupérer le scaler des métadonnées
                self.scalers['tensorflow'] = self.metadata['tensorflow']['scaler']
                
                print("Modèle TensorFlow chargé avec succès.")
            else:
                print("Modèle TensorFlow ou ses fichiers associés non trouvés.")
        except Exception as e:
            print(f"Erreur lors du chargement du modèle TensorFlow: {e}")
        
        # TensorFlow Lite
        try:
            tflite_path = os.path.join('tf_model', 'model.tflite')
            
            if os.path.exists(tflite_path):
                # Charger le modèle TFLite
                self.models['tensorflow_lite'] = tflite_path
                print("Modèle TensorFlow Lite chargé avec succès.")
            else:
                print("Modèle TensorFlow Lite non trouvé.")
        except Exception as e:
            print(f"Erreur lors du chargement du modèle TensorFlow Lite: {e}")
    
    def predict_crop_sklearn(self, N, P, K, temperature, humidity, ph, rainfall):
        """
        Prédit la culture en utilisant le modèle scikit-learn
        """
        if self.models['gradient_boosting'] is None or self.scalers['gradient_boosting'] is None:
            raise ValueError("Le modèle Gradient Boosting n'est pas chargé")
        
        # Créer un tableau avec les valeurs d'entrée
        input_data = np.array([[N, P, K, temperature, humidity, ph, rainfall]])
        
        # Standardiser les données d'entrée
        input_data_scaled = self.scalers['gradient_boosting'].transform(input_data)
        
        # Faire la prédiction
        prediction = self.models['gradient_boosting'].predict(input_data_scaled)
        probabilities = self.models['gradient_boosting'].predict_proba(input_data_scaled)
        
        # Trouver l'indice de la classe prédite
        predicted_idx = np.argmax(probabilities[0])
        confidence = probabilities[0][predicted_idx]
        
        return prediction[0], confidence
    
    def predict_crop_tensorflow(self, N, P, K, temperature, humidity, ph, rainfall):
        """
        Prédit la culture en utilisant le modèle TensorFlow
        """
        if self.models['tensorflow'] is None or self.scalers['tensorflow'] is None:
            raise ValueError("Le modèle TensorFlow n'est pas chargé")
        
        # Créer un tableau avec les valeurs d'entrée
        input_data = np.array([[N, P, K, temperature, humidity, ph, rainfall]])
        
        # Standardiser les données d'entrée
        input_data_scaled = self.scalers['tensorflow'].transform(input_data)
        
        # Faire la prédiction
        prediction_proba = self.models['tensorflow'].predict(input_data_scaled)
        prediction_idx = np.argmax(prediction_proba[0])
        
        # Convertir l'indice en nom de classe
        predicted_crop = self.metadata['tensorflow']['class_names'][prediction_idx]
        confidence = prediction_proba[0][prediction_idx]
        
        return predicted_crop, confidence
    
    def predict_crop_tflite(self, N, P, K, temperature, humidity, ph, rainfall):
        """
        Prédit la culture en utilisant le modèle TensorFlow Lite
        """
        if self.models['tensorflow_lite'] is None or self.scalers['tensorflow'] is None:
            raise ValueError("Le modèle TensorFlow Lite n'est pas chargé")
        
        # Charger l'interpréteur TFLite
        interpreter = tf.lite.Interpreter(model_path=self.models['tensorflow_lite'])
        interpreter.allocate_tensors()
        
        # Obtenir les détails des entrées et sorties
        input_details = interpreter.get_input_details()
        output_details = interpreter.get_output_details()
        
        # Créer un tableau avec les valeurs d'entrée
        input_data = np.array([[N, P, K, temperature, humidity, ph, rainfall]])
        
        # Standardiser les données d'entrée
        input_data_scaled = self.scalers['tensorflow'].transform(input_data).astype(np.float32)
        
        # Définir les données d'entrée
        interpreter.set_tensor(input_details[0]['index'], input_data_scaled)
        
        # Exécuter l'inférence
        interpreter.invoke()
        
        # Obtenir les résultats
        prediction_proba = interpreter.get_tensor(output_details[0]['index'])
        prediction_idx = np.argmax(prediction_proba[0])
        
        # Convertir l'indice en nom de classe
        predicted_crop = self.metadata['tensorflow']['class_names'][prediction_idx]
        confidence = prediction_proba[0][prediction_idx]
        
        return predicted_crop, confidence
    
    def predict_crop(self, N, P, K, temperature, humidity, ph, rainfall, model_type='gradient_boosting'):
        """
        Prédit la culture en utilisant le modèle spécifié
        """
        if model_type == 'gradient_boosting':
            return self.predict_crop_sklearn(N, P, K, temperature, humidity, ph, rainfall)
        elif model_type == 'tensorflow':
            return self.predict_crop_tensorflow(N, P, K, temperature, humidity, ph, rainfall)
        elif model_type == 'tensorflow_lite':
            return self.predict_crop_tflite(N, P, K, temperature, humidity, ph, rainfall)
        else:
            raise ValueError(f"Type de modèle inconnu: {model_type}")

# Fonction principale pour démontrer l'utilisation
def main():
    # Créer une instance du prédicteur
    predictor = CropPredictor()
    
    # Paramètres de test
    N = 90
    P = 42
    K = 43
    temperature = 21
    humidity = 82
    ph = 6.5
    rainfall = 200
    
    print(f"\nPour les valeurs d'entrée:\nN={N}, P={P}, K={K}, température={temperature}, humidité={humidity}, pH={ph}, précipitations={rainfall}")
    
    # Prédiction avec les différents modèles
    try:
        crop_gb, confidence_gb = predictor.predict_crop(N, P, K, temperature, humidity, ph, rainfall, model_type='gradient_boosting')
        print(f"\nPrédiction avec Gradient Boosting: {crop_gb} (confiance: {confidence_gb:.4f})")
    except Exception as e:
        print(f"Erreur avec Gradient Boosting: {e}")
    
    try:
        crop_tf, confidence_tf = predictor.predict_crop(N, P, K, temperature, humidity, ph, rainfall, model_type='tensorflow')
        print(f"Prédiction avec TensorFlow: {crop_tf} (confiance: {confidence_tf:.4f})")
    except Exception as e:
        print(f"Erreur avec TensorFlow: {e}")
    
    try:
        crop_tflite, confidence_tflite = predictor.predict_crop(N, P, K, temperature, humidity, ph, rainfall, model_type='tensorflow_lite')
        print(f"Prédiction avec TensorFlow Lite: {crop_tflite} (confiance: {confidence_tflite:.4f})")
    except Exception as e:
        print(f"Erreur avec TensorFlow Lite: {e}")
    
    # Afficher les seuils de confiance avec un graphique
    models = []
    confidences = []
    
    try:
        models.append('Gradient Boosting')
        confidences.append(confidence_gb)
    except:
        pass
    
    try:
        models.append('TensorFlow')
        confidences.append(confidence_tf)
    except:
        pass
    
    try:
        models.append('TensorFlow Lite')
        confidences.append(confidence_tflite)
    except:
        pass
    
    if models and confidences:
        plt.figure(figsize=(10, 6))
        sns.barplot(x=models, y=confidences)
        plt.title('Confiance des prédictions par modèle')
        plt.ylabel('Confiance')
        plt.ylim(0, 1)
        plt.savefig('model_confidence.png')
        plt.show()

# Interface utilisateur simple pour tester les prédictions
def predict_with_user_input():
    predictor = CropPredictor()
    
    print("\n===== Prédiction de cultures agricoles =====")
    print("Veuillez entrer les valeurs suivantes :")
    
    try:
        N = float(input("Niveau d'azote (N) : "))
        P = float(input("Niveau de phosphore (P) : "))
        K = float(input("Niveau de potassium (K) : "))
        temperature = float(input("Température (°C) : "))
        humidity = float(input("Humidité (%) : "))
        ph = float(input("pH du sol : "))
        rainfall = float(input("Précipitations (mm) : "))
        
        # Sélectionner le modèle
        print("\nModèles disponibles:")
        print("1. Gradient Boosting")
        print("2. TensorFlow")
        print("3. TensorFlow Lite")
        model_choice = int(input("Choisissez un modèle (1-3) : "))
        
        model_type = {
            1: 'gradient_boosting',
            2: 'tensorflow',
            3: 'tensorflow_lite'
        }.get(model_choice, 'gradient_boosting')
        
        # Faire la prédiction
        crop, confidence = predictor.predict_crop(N, P, K, temperature, humidity, ph, rainfall, model_type)
        
        print(f"\nRésultat de la prédiction :")
        print(f"Culture recommandée : {crop}")
        print(f"Niveau de confiance : {confidence:.4f} ({confidence*100:.2f}%)")
        
    except ValueError as e:
        print(f"Erreur : {e}. Assurez-vous d'entrer des valeurs numériques valides.")
    except Exception as e:
        print(f"Une erreur s'est produite : {e}")

# Afficher les caractéristiques requises pour certaines cultures
def show_crop_requirements():
    try:
        # Charger le dataset original pour analyser les besoins des cultures
        df = pd.read_csv("Crop_recommendation.csv")
        
        # Grouper par culture et calculer les moyennes
        crop_requirements = df.groupby('label').agg({
            'N': 'mean',
            'P': 'mean',
            'K': 'mean',
            'temperature': 'mean',
            'humidity': 'mean',
            'ph': 'mean',
            'rainfall': 'mean'
        }).round(2)
        
        print("\n===== Besoins moyens par culture =====")
        print(crop_requirements)
        
        # Créer des graphiques pour visualiser les besoins
        # Température et humidité
        plt.figure(figsize=(12, 8))
        plt.scatter(crop_requirements['temperature'], crop_requirements['humidity'], s=100)
        
        # Ajouter des étiquettes pour chaque point
        for i, crop in enumerate(crop_requirements.index):
            plt.annotate(crop, 
                         (crop_requirements['temperature'][i], crop_requirements['humidity'][i]),
                         xytext=(5, 5), textcoords='offset points')
            
        plt.title('Besoins en température et humidité par culture')
        plt.xlabel('Température moyenne (°C)')
        plt.ylabel('Humidité moyenne (%)')
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.savefig('temp_humidity_requirements.png')
        plt.show()
        
        # Nutriments N, P, K
        plt.figure(figsize=(12, 8))
        
        x = np.arange(len(crop_requirements.index))
        width = 0.25
        
        plt.bar(x - width, crop_requirements['N'], width, label='Azote (N)')
        plt.bar(x, crop_requirements['P'], width, label='Phosphore (P)')
        plt.bar(x + width, crop_requirements['K'], width, label='Potassium (K)')
        
        plt.xlabel('Culture')
        plt.ylabel('Niveau moyen')
        plt.title('Besoins en nutriments (N, P, K) par culture')
        plt.xticks(x, crop_requirements.index, rotation=90)
        plt.legend()
        plt.tight_layout()
        plt.savefig('npk_requirements.png')
        plt.show()
        
        return crop_requirements
        
    except Exception as e:
        print(f"Erreur lors de l'affichage des besoins des cultures: {e}")
        return None

# Menu principal
def menu():
    while True:
        print("\n===== AgroPredict: Système de recommandation de cultures =====")
        print("1. Faire une prédiction avec des valeurs par défaut")
        print("2. Faire une prédiction avec mes propres valeurs")
        print("3. Afficher les besoins moyens par culture")
        print("4. Quitter")
        
        try:
            choice = int(input("Choisissez une option (1-4) : "))
            
            if choice == 1:
                main()
            elif choice == 2:
                predict_with_user_input()
            elif choice == 3:
                show_crop_requirements()
            elif choice == 4:
                print("Merci d'avoir utilisé AgroPredict. Au revoir!")
                break
            else:
                print("Option invalide. Veuillez choisir une option entre 1 et 4.")
        except ValueError:
            print("Entrée invalide. Veuillez entrer un nombre entre 1 et 4.")
        except Exception as e:
            print(f"Une erreur s'est produite : {e}")

# Exécuter le menu si le script est lancé directement
if __name__ == "__main__":
    menu()
```
