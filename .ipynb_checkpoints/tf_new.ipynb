```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import os
```

```python
# Vérifier la version de TensorFlow
print(f"TensorFlow version: {tf.__version__}")

# Chargement des données
df = pd.read_csv("../Datasets/Crop_recommendation.csv")
print("Forme du dataset:", df.shape)

# Affichage des premières lignes
print("\nAperçu des données:")
print(df.head())
```

```python
# Préparation des données
X = df.iloc[:, :-1].values  # Toutes les colonnes sauf la dernière (label)
y = df.iloc[:, -1].values   # Dernière colonne (label)

# Encoder les labels catégoriels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Obtenir les classes uniques pour référence future
class_names = label_encoder.classes_
num_classes = len(class_names)
print(f"\nNombre de classes: {num_classes}")
print(f"Classes: {class_names}")

# Standardisation des caractéristiques
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Séparation en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# Convertir en format spécifique à TensorFlow
X_train = np.array(X_train, dtype=np.float32)
X_test = np.array(X_test, dtype=np.float32)
y_train = np.array(y_train, dtype=np.int32)
y_test = np.array(y_test, dtype=np.int32)
```

```python
# Convertir les labels en catégories one-hot
y_train_one_hot = tf.keras.utils.to_categorical(y_train, num_classes=num_classes)
y_test_one_hot = tf.keras.utils.to_categorical(y_test, num_classes=num_classes)
```

```python
# Créer un modèle Gradient Boosting à l'aide de TensorFlow
def create_model(num_features, num_classes):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(128, activation='relu', input_shape=(num_features,)),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.1),
        tf.keras.layers.Dense(num_classes, activation='softmax')
    ])
     
    # Compiler le modèle
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model
    
```

```python
# Créer le modèle
model = create_model(X_train.shape[1], num_classes)

# Résumé du modèle
print("\nStructure du modèle:")
model.summary()

# Définir les callbacks
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=5,
    min_lr=0.0001
)

# Créer un dossier pour les modèles
model_dir = 'tf_model'
os.makedirs(model_dir, exist_ok=True)

# Checkpoint pour sauvegarder le meilleur modèle
model_checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath=os.path.join(model_dir, 'best_model.h5'),
    monitor='val_accuracy',
    save_best_only=True,
    verbose=1
)

# Entraîner le modèle
epochs = 50
batch_size = 32

history = model.fit(
    X_train, y_train_one_hot,
    epochs=epochs,
    batch_size=batch_size,
    validation_data=(X_test, y_test_one_hot),
    callbacks=[early_stopping, reduce_lr, model_checkpoint],
    verbose=1
)
```

```python
# Évaluer le modèle
test_loss, test_accuracy = model.evaluate(X_test, y_test_one_hot, verbose=0)
print(f"\nPrécision sur l'ensemble de test: {test_accuracy:.4f}")
```

```python
# Visualiser l'historique d'entraînement
plt.figure(figsize=(12, 5))

# Précision
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Entraînement')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Précision du modèle')
plt.xlabel('Epoch')
plt.ylabel('Précision')
plt.legend()
```

```python
# Perte
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Entraînement')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Perte du modèle')
plt.xlabel('Epoch')
plt.ylabel('Perte')
plt.legend()

plt.tight_layout()
plt.savefig('tf_training_history.png')
plt.show()
```

```python
# Faire des prédictions
y_pred_proba = model.predict(X_test)
y_pred = np.argmax(y_pred_proba, axis=1)

# Calculer la précision
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("\nRapport de classification:")
print(classification_report(y_test, y_pred, target_names=class_names))
```

```python
# Matrice de confusion
plt.figure(figsize=(14, 12))
cm = confusion_matrix(y_test, y_pred)
import seaborn as sns
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Matrice de confusion')
plt.xlabel('Prédictions')
plt.ylabel('Valeurs réelles')
plt.tight_layout()
plt.savefig('tf_confusion_matrix.png')
plt.show()
```

```python
# Fonction de prédiction pour de nouvelles données
def predict_crop_tf(N, P, K, temperature, humidity, ph, rainfall):
    # Créer un tableau avec les valeurs d'entrée
    input_data = np.array([[N, P, K, temperature, humidity, ph, rainfall]])
    
    # Standardiser les données d'entrée
    input_data_scaled = scaler.transform(input_data)
    
    # Faire la prédiction
    prediction_proba = model.predict(input_data_scaled)
    prediction_idx = np.argmax(prediction_proba)
    
    # Convertir l'indice en nom de classe
    predicted_crop = class_names[prediction_idx]
    confidence = prediction_proba[0][prediction_idx]
    
    return predicted_crop, confidence

# Exemple de prédiction
N = 90
P = 42
K = 43
temperature = 21
humidity = 82
ph = 6.5
rainfall = 200

predicted_crop, confidence = predict_crop_tf(N, P, K, temperature, humidity, ph, rainfall)
print(f"\nPour les valeurs d'entrée N={N}, P={P}, K={K}, température={temperature}, humidité={humidity}, pH={ph}, précipitations={rainfall}")
print(f"La culture prédite est: {predicted_crop} avec une confiance de {confidence:.4f}")

# Sauvegarder les métadonnées
import pickle

metadata = {
    'feature_names': list(df.columns[:-1]),
    'label_encoder': label_encoder,
    'scaler': scaler,
    'num_classes': num_classes,
    'class_names': class_names.tolist()
}

with open(os.path.join(model_dir, 'metadata.pkl'), 'wb') as f:
    pickle.dump(metadata, f)

# Convertir le modèle TensorFlow en TensorFlow Lite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Sauvegarder le modèle TFLite
with open(os.path.join(model_dir, 'model.tflite'), 'wb') as f:
    f.write(tflite_model)

print(f"\nModèle TensorFlow, modèle TFLite et métadonnées sauvegardés dans le dossier '{model_dir}'")
```
